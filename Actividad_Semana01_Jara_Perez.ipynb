### Pregunta 1: Listas y Aleatorios

Crear una lista llamada `lst_numeros` con 300 numeros aleatorios entre 100 y 600.
Luego:
1. Muestra el 25% de los primeros numeros generados (usa indexacion de listas)
2. Muestra cuantos numeros son mayores a 400
3. Muestra el valor maximo y minimo de la lista

**Tip:** Usa `random.randint()` o list comprehension con `random`

````python
import random

random.seed(42)  # Para reproducibilidad

# Tu codigo aqui
lst_numeros = [random.randint(100, 600) for _ in range(300)]

# Muestra el 25% de los primeros elementos de la lista utilizando slicing y convirtiendo el resultado a entero 
# utilizando la función int() para obtener el índice correcto

lst_mod = lst_numeros[:int(len(lst_numeros) * 0.25)]

# Filtra aquellos numeros mayores a 400 utilizando una comprensión de listas

lst_mayores = [numero for numero in lst_numeros if numero > 400]

# Crea una lista con el número máximo y mínimo de la lista original utilizando las funciones max() y min()
lst_mayor_min = [max(lst_numeros), min(lst_numeros)]

print("Mostrar 25% de los primeros elementos:", lst_mod)
print("Cantidad de numeros mayores a 400:", len(lst_mayores))
print("Número máximo y mínimo:", lst_mayor_min)
````

### Pregunta 2: Funciones y Formula Cuadratica

Crea una **funcion** llamada `formula_cuadratica(a, b, c)` que retorne las dos raices de la ecuacion cuadratica:

$$x_{1,2} = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}$$

Pruebala con: `a=4, b=9, c=3`

**Bonus:** Agrega validacion para el caso donde el discriminante es negativo (raices complejas)

````python
# Tu codigo aqui
def formula_cuadratica(a,b,c):

    # Calcula el discriminante
    discriminante = b**2 - 4*a*c

    #Valida el discriminante para determinar el número de soluciones reales
    if discriminante < 0:
        return "No hay soluciones reales"
    elif discriminante == 0:
        x = -b / (2*a)
        return f"Una solución real: x = {x}"
    else:
        x1 = (-b + discriminante**0.5) / (2*a)
        x2 = (-b - discriminante**0.5) / (2*a)
        return f"Dos soluciones reales: x1 = {x1}, x2 = {x2}"
    

print(formula_cuadratica(4, 9, 3))      
````

### Pregunta 4: List Comprehension

Dada la siguiente lista de temperaturas en Celsius:
```python
celsius = [0, 10, 20, 25, 30, 35, 40]
```

Usando **list comprehension**:
1. Convierte todas las temperaturas a Fahrenheit: `F = C * 9/5 + 32`
2. Filtra solo las temperaturas en Celsius mayores a 25
3. Crea una lista de tuplas `(celsius, fahrenheit)` para cada temperatura

````python
celsius = [0, 10, 20, 25, 30, 35, 40]

# Tu codigo aqui
farenheit = [int(((temp * 1.8) + 32)) for temp in celsius]
filtrar_temp = [temp for temp in celsius if temp > 25]

tupla_celsius = tuple(celsius)
tupla_farenheit = tuple(farenheit)

print("Temperaturas en Farenheit:", farenheit)
print("Temperaturas mayores a 25°C:", filtrar_temp)

print ("Tupla de Celsius:", tupla_celsius)
print ("Tupla de Farenheit:", tupla_farenheit)

print (tupla_celsius + tupla_farenheit)
````

### Pregunta 5: Funciones Lambda + Map

Dada la siguiente lista de precios en dolares:
```python
precios_usd = [10.5, 25.0, 8.75, 42.0, 15.3]
```

1. Usa `map` con una funcion `lambda` para convertir todos los precios a soles (tipo de cambio = 3.75)
2. Usa `filter` con `lambda` para obtener solo los precios en soles mayores a 50
3. Muestra los resultados con f-strings formateados a 2 decimales

````python
precios_usd = [10.5, 25.0, 8.75, 42.0, 15.3]

# Tu codigo aqui
# FUNCION QUE SE ENCARGA DE REDONDEAR LOS PRECIOS A 2 DECIMALES DE LAS LISTAS
def redondearListas(lista):
    return list(map(lambda precio: round(precio, 2), lista))

# Convierte los precios de USD a PEN utilizando map() y una función lambda que multiplica cada precio por 3.75
#map devuelve un objeto iterable, por lo que se convierte a una lista utilizando list() para obtener los precios en PEN como una lista.
#1
precios_pen = list(map(lambda precio: precio * 3.75, precios_usd))
#2
precios_mayores_50 = list(filter(lambda precio: precio > 50, precios_pen))
#3
precios_redondeados = redondearListas(precios_pen)
precios_redondeados2 = redondearListas(precios_mayores_50)

print (f"Precios en PEN: {precios_redondeados}")
print (f"Precios mayores a 50 PEN: {precios_redondeados2}")
````

### Pregunta 6: Estadisticos con NumPy

Genera datos simulados de 100 estudiantes:
- `horas_estudio`: enteros aleatorios entre 1 y 12
- `notas`: numeros aleatorios con distribucion normal, media=14, desviacion=3

Usa `np.random.seed(42)` para reproducibilidad.

Luego calcula:
1. Media, mediana y desviacion estandar de ambas variables
2. El coeficiente de correlacion entre horas_estudio y notas
3. Cuantos estudiantes tienen nota >= 11 (aprobados)
4. El promedio de horas de estudio de los aprobados vs los desaprobados

Muestra todos los resultados con f-strings.

````python
import numpy as np
np.random.seed(42)

# Tu codigo aqui
horas_estudio = np.random.randint(1, 12, 100)
notas = np.random.normal(loc=14, scale=3, size=100)

# Calcula la media, mediana y desviación estándar de ambas listas

mediana = np.median(horas_estudio)
media = np.mean(horas_estudio)
desviacion_estandar = np.std(horas_estudio)

# Coeficiente de correlación entre horas de estudio y notas 

coef = np.corrcoef(horas_estudio, notas)[0, 1]
print(coef)

#Saber cuantos estudiantes tienen nota mayor o igual a 11

estudiantes_aprobados = np.sum(notas >= 11)
print(f"Cantidad de estudiantes con nota mayor o igual a 11: {estudiantes_aprobados}")

#PROMEDIO DE HORAS DE ESTUDIO DE LOS ESTUDIANTES APROBADOS Y DESAPROBADOS

horas_aprobados = horas_estudio[notas >= 11]
horas_desaprobados = horas_estudio[notas < 11]

promedio_horas_aprobados = np.mean(horas_aprobados)
promedio_horas_desaprobados = np.mean(horas_desaprobados)

print(f"Promedio de horas de estudio de estudiantes aprobados: {promedio_horas_aprobados}")
print(f"Promedio de horas de estudio de estudiantes desaprobados: {promedio_horas_desaprobados}")
````

### Pregunta 7: Manipulacion de Arrays

1. Crea un array de 60 numeros equidistantes entre 0 y 100 usando `np.linspace`
2. Reorganizalo en una matriz de 10 filas x 6 columnas con `reshape`
3. Muestra la forma del array resultante
4. Calcula el promedio de cada columna (usa `axis=0`)
5. Calcula el promedio de cada fila (usa `axis=1`)
6. Selecciona las filas donde el valor de la primera columna es mayor a 50
7. Transpone la matriz y muestra su nueva forma

````python
import numpy as np

# Tu codigo aqui
def convertir_2_decimales(array):
    return np.round(array, 2)

#1 
array = np.linspace(start=0, stop=100, num=60)
print("Array con 2 decimales:", convertir_2_decimales(array))

#2
matriz = array.reshape(10,6)

#3
print("Matriz de 10 filas y 6 columnas:", convertir_2_decimales(matriz))

#4
calculo_prom_columna = np.mean(matriz, axis=0)
print("Promedio de cada columna:", convertir_2_decimales(calculo_prom_columna))

#5
calculo_prom_fila = np.mean(matriz, axis=1)
print("Promedio de cada fila:", convertir_2_decimales(calculo_prom_fila))

#6
#DAME LA MATRIZ FILTRADA DONDE EL PRIMER ELEMENTO DE CADA COLUMNA SEA MAYOR A 50
#Filtra las filas donde la columna 0 cumpla la condición
nueva_matriz = matriz[matriz[:, 0] > 50]
print("Filas donde el primer elemento es mayor a 50:",  convertir_2_decimales(nueva_matriz))

#7
matriz_transpuesta = matriz.T
print('Matriz transpuesta: ', convertir_2_decimales(matriz_transpuesta))

````
